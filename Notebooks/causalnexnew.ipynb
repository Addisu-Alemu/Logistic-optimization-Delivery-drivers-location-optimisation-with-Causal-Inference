```python
from geopy import distance
import pandas as pd

def calculate_distances(starting_coordinates, ending_coordinates):
    calculated_distances = []
    for i in range(len(starting_coordinates)):
        # Handle cases where coordinates are not properly formatted
        try:
            val = str(starting_coordinates[i]).split(',')
            starting_tuple = (float(val[0]), float(val[1]))  # Convert to float
            val_end = str(ending_coordinates[i]).split(',')
            ending_tuple = (float(val_end[0]), float(val_end[1]))  # Convert to float
            calculated_distances.append(distance.distance(starting_tuple, ending_tuple).km)
        except (IndexError, ValueError):
            print(f"Skipping calculation for index {i}: Invalid coordinates")
            calculated_distances.append(None)  # Or append a default value
    return calculated_distances


```

```python
drop_col = ['order_id','driver_id','lat','lng','trip_origin','trip_destination','trip_start_date','id' ]
data = df.drop(columns=drop_col)
data.head(5)
```

```python
# Replace 'rejected' with 0 and 'accepted' with 1 in the 'driver_action' column
data['driver_action'] = data['driver_action'].replace({'rejected': 0, 'accepted': 1})

# Save the modified DataFrame to a new CSV file
data.to_csv("../data/structured_data.csv", index=False)
```

```python
data = pd.read_csv("../data/structured_data.csv")
data.head(5)
```

```python
# choose non-numeric columns
import numpy as np

struct_data = data.copy()
non_numeric_columns = list(struct_data.select_dtypes(exclude=[np.number]).columns)

print(non_numeric_columns)
```

```python
import pandas as pd
import numpy as np

# Impute NaN with column means
for col in struct_data.columns:
    if struct_data[col].isnull().any():
        mean_value = struct_data[col].mean()
        struct_data[col].fillna(mean_value, inplace=True)
```

```python
#apply the NOTEARS algorithm to learn the structure.

from causalnex.structure.notears import from_pandas
sm = from_pandas(struct_data)
```

```python
from causalnex.plots import plot_structure, NODE_STYLE, EDGE_STYLE
viz = plot_structure(
    sm,
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
)

viz.toggle_physics(False)
viz.show("../screenshots/fully_connected.html")
```
